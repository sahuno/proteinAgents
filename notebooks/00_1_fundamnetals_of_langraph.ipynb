## Here, we explore TypedDict
### definition, running

```python
from typing_extensions import TypedDict

# TypedDict: defines the schema for the nodes
class State(TypedDict):
    graph_state: str
```

```python
def node_1(state):
    print("---Node 1---")
    return {"graph_state":  "I am " + state['graph_state']}

def node_2(state):
    print("---Node 2---")
    return {"graph_state": state['graph_state'] +" happy!"}

def node_3(state):
    print("---Node 3---")
    return {"graph_state": state['graph_state'] +" sad!"}
```
